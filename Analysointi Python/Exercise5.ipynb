```python
# getting items which User Rating is lower than 45
subset_df = chipo[chipo['User Rating'] > 45]
print(subset_df)

# subsetting based on text data
subset_df = chipo[chipo['Game Title'] == ""]
print(subset_df)

# subsetting with mulitple conditions
subset_df1 = chipo['User Rating'] > 45
subset_df2 = chipo['Game Title'] == ""
subset_df3 = chipo[subset_df1 & subset_df2]
subset_df3.head()
```

```python
# if you check the data type for item_price column it is object and therefore we need to change the data type to float or integer to perform calculations
print(chipo['Price'].dtype)
```

```python
# Now lets make a new column called total by multiplying 'quantity' and 'item_price'
chipo['Story Quality'] = chipo['Game Length (Hours)'] * chipo['Price']

# lets check if we have the new column total
chipo.head(20)
```

```python
# First, let's create a function to define the price ranges
def categorize_price(Price):  # Use 'item_price' to match the column name
    if Price < 30:
        return 'Low'
    elif Price < 45:
        return 'Medium'
    else:
        return 'High'

# We can use the .apply() method to create a new column 'price_group' based on the 'item_price' column.
chipo['price_group'] = chipo['Price'].apply(categorize_price)
chipo.head()
```
